```python
import pandas as pd
```

```python
# glucose_in_blood, white_blood_cells, billirubin_in_plasma
query_name = 'bilirubin_in_plasma'

# read csv
df_0 = pd.read_csv(f'{query_name}/{query_name}0.csv')
df_1 = pd.read_csv(f'{query_name}/{query_name}1.csv')
df_2 = pd.read_csv(f'{query_name}/{query_name}2.csv')

column_names = ['loinc_num'.upper(), 'long_common_name'.upper(), 'component'.upper(), 'system'.upper(), 'property'.upper(), 'rank'.upper()]
# delete the columns not in list
for names in df_0.columns:
    if names.upper() not in column_names:
        df_0.drop(names, axis=1, inplace=True)
        df_1.drop(names, axis=1, inplace=True)
        df_2.drop(names, axis=1, inplace=True)

df_0['RANK'] = 0
df_1['RANK'] = 1
df_2['RANK'] = 2
```

```python
df_ext = df_2.append(df_1).drop_duplicates(subset=['LOINC_NUM'], keep='first')
df_ext = df_ext.append(df_0).drop_duplicates(subset=['LOINC_NUM'], keep='first')
df_ext
```

```python
if query_name == 'glucose_in_blood': sheet_name = 0
if query_name == 'white_blood_cells': sheet_name = 1
if query_name == 'billirubin_in_plasma': sheet_name = 2

#read excel
df_loinc = pd.read_excel('loinc.xlsx', sheet_name = sheet_name)
df_loinc = df_loinc[2:]
# loinc_num	long_common_name	component	system	property	rank
column_names = ['loinc_num'.upper(), 'long_common_name'.upper(), 'component'.upper(), 'system'.upper(), 'property'.upper(), 'rank'.upper()]

# rename columns
df_loinc.columns = column_names
```

```python
df_final = df_loinc.append(df_ext).drop_duplicates(subset=['LOINC_NUM'], keep='last')
df_final['RANK'].fillna(0, inplace=True)
df_final
```

```python
df_final = pd.merge(df_final, df_loinc, how='inner', on=['LOINC_NUM'])
df_final.drop(columns=['LONG_COMMON_NAME_y', 'COMPONENT_y','SYSTEM_y','PROPERTY_y','RANK_y'],inplace=True)
df_final.columns = column_names
df_final
```

```python
df_0=df_0[0:50]
df_1=df_1[0:50]
df_2=df_2[0:50]
df_ext = df_2.append(df_1).drop_duplicates(subset=['LOINC_NUM'], keep='first')
df_ext = df_ext.append(df_0).drop_duplicates(subset=['LOINC_NUM'], keep='first')
df_final = df_final.append(df_ext).drop_duplicates(subset=['LOINC_NUM'], keep='first')
df_final
```

```python
# save to excel
df_final.to_excel(f'{query_name}_expanded.xlsx', index=False)
```
